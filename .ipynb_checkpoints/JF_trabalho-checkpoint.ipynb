## Data Cleaning

```python
df_clean = df.copy()
print(f"Initial shape: {df_clean.shape}")
```

```python
def _drop_columns(dfx, cols, reason):
    if cols:
        print(f"\nDropped {len(cols)} column(s) — {reason}:")
        for c in cols:
            print(f"  • {c}")
        dfx.drop(columns=cols, inplace=True)
    else:
        print(f"\nNo columns dropped — {reason}.")
    return dfx
```

```python
# 1) Type coercion (example: focus_factor → numeric)
if "focus_factor" in df_clean.columns and not pd.api.types.is_numeric_dtype(df_clean["focus_factor"]):
    df_clean["focus_factor"] = pd.to_numeric(df_clean["focus_factor"], errors="coerce")
    print("Converted 'focus_factor' to numeric (non-convertible → NaN)")
```

```python
# 2) Replace +/- infinito with NaN (vectorized for all numeric columns)
num_cols = df_clean.select_dtypes(include="number").columns
n_inf_before = np.isinf(df_clean[num_cols].to_numpy()).sum()
if n_inf_before:
    df_clean[num_cols] = df_clean[num_cols].replace([np.inf, -np.inf], np.nan)
    print(f"Replaced {n_inf_before} ±inf value(s) with NaN across numeric columns.")
else:
    print("No +/- inf values found in numeric columns.")
```

```python
# 3) Drop entirely empty columns (all NaN)
empty_cols = [c for c in df_clean.columns if df_clean[c].isna().all()]
df_clean = _drop_columns(df_clean, empty_cols, "completely empty (all NaN)")
print(f"Shape after dropping empty columns: {df_clean.shape}")
```

```python
# 4) Remove duplicate rows
n_dups = int(df_clean.duplicated().sum())
if n_dups:
    df_clean = df_clean.drop_duplicates().reset_index(drop=True)
    print(f"\nRemoved {n_dups} duplicate row(s).")
else:
    print("\nNo duplicate rows found.")
print(f"Shape after removing duplicates: {df_clean.shape}")
```

```python
# 5) Confirmar colunas constantes (sem variabilidade)
const_cols = [c for c in df_clean.columns 
              if c not in ["target_class", "target_regression"] 
              and df_clean[c].nunique(dropna=False) <= 1]
if const_cols:
    print(f"\nDropped constant columns: {const_cols}")
    df_clean.drop(columns=const_cols, inplace=True)
else:
    print("\nNo constant columns found.")
```

```python
# 6) Missing-value summary (only columns that have NaN)
missing_counts = df_clean.isna().sum()
missing_cols = missing_counts[missing_counts > 0].sort_values(ascending=False)
if not missing_cols.empty:
    print("\nColumns containing missing values (count | %):")
    for col, n_miss in missing_cols.items():
        pct = 100 * n_miss / len(df_clean)
        print(f"  • {col}: {n_miss} ({pct:.2f}%)")
else:
    print("\nNo missing values found in any column.")

print(f"\nFinal shape after cleaning: {df_clean.shape}")
```

```python
# 7) Handle missing vallues

num_cols = df_clean.select_dtypes(include="number").columns
cols_with_na = [c for c in num_cols if df_clean[c].isna().any()]

if cols_with_na:
    print(f"\nReplacing NaN values with column mean for {len(cols_with_na)} numeric column(s):")
    total_replaced = 0

    for c in cols_with_na:
        n_missing = int(df_clean[c].isna().sum())
        mean_val = df_clean[c].mean()
        df_clean[c].fillna(mean_val, inplace=True)
        total_replaced += n_missing
        print(f"  • {c}: {n_missing} value(s) replaced with mean = {mean_val:.4f}")

    print(f"\nTotal NaN values replaced: {total_replaced}")
else:
    print("\nNo numeric columns contained NaN values — no replacements performed.")

# Verificação final
remaining_nans = df_clean.isna().sum().sum()
if remaining_nans == 0:
    print("No remaining NaN values after replacement.")
else:
    print(f"{remaining_nans} NaN values remain (possibly in non-numeric columns).")

print(f"Shape after missing-value treatment: {df_clean.shape}")
```

```python
# 8) Categóricas: preencher NaN com a moda e normalizar texto
cat_cols = df_clean.select_dtypes(exclude="number").columns.tolist()

from sklearn.impute import SimpleImputer
if cat_cols:
    cat_imp = SimpleImputer(strategy="most_frequent")
    df_clean[cat_cols] = cat_imp.fit_transform(df_clean[cat_cols])

    # Normalizar strings (lowercase, trim)
    for c in cat_cols:
        if df_clean[c].dtype == "object":
            df_clean[c] = df_clean[c].astype(str).str.strip().str.lower()
print(" Categóricas imputadas e normalizadas.")
```

```python
# 9) Tratamento de booleanos
bool_like = [c for c in df_clean.columns 
             if df_clean[c].nunique(dropna=True) <= 3 and df_clean[c].dtype != "float64"]

for c in bool_like:
    df_clean[c] = df_clean[c].replace(
        {"true":1, "false":0, "yes":1, "no":0}
    )
    # tentar coerção para 0/1
    df_clean[c] = pd.to_numeric(df_clean[c], errors="coerce")
    # qualquer coisa não 0 torna-se 1 (após imputação de NaN para 0)
    df_clean[c] = df_clean[c].fillna(0).astype(int).clip(0,1)
print("Flags/booleanos coerentes em 0/1.")
```

```python
# 10 Tratamento de Outliers 

num_cols = df_clean.select_dtypes(include="number").columns.tolist()
for c in num_cols:
    q1, q3 = df_clean[c].quantile([0.25, 0.75])
    iqr = q3 - q1
    low, high = q1 - 1.5*iqr, q3 + 1.5*iqr
    df_clean[c] = df_clean[c].clip(lower=low, upper=high)
print("Outliers atenuados via IQR clipping (opcional).")
```

```python
# Validar targets (sem modificar os dados)
if "target_class" in df_clean.columns:
    print("\nTarget class distribution (class imbalance check):")
    display(df_clean["target_class"].value_counts())

if "target_regression" in df_clean.columns:
    print("\nRegression target summary:")
    display(df_clean["target_regression"].describe())

# Verificação final
print("\nFinal cleaning summary:")
print(f"  - Rows: {df_clean.shape[0]}")
print(f"  - Columns: {df_clean.shape[1]}")
print(f"  - Missing values remaining: {df_clean.isna().sum().sum()}")
print(f"  - Duplicates remaining: {df_clean.duplicated().sum()}")
```
