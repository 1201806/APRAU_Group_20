#### Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dataset.csv")

# Configurações visuais
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('viridis')
```

#### Visualização inicial

```python
print("Dimensões do dataset:", df.shape)
display(df.head())

df.info()
```

#### Verificação Valores nulos 

```python
display(df.describe().T)

# Verificar valores nulos
print("\nValores em falta por coluna:")
print(df.isnull().sum())
```

#### Univariate Analysis 

```python
numeric_features = df.select_dtypes(include=np.number).columns

df[numeric_features].hist(figsize=(15, 12), bins=20)
plt.suptitle("Distribuição das variáveis numéricas", fontsize=16)
plt.show()

# Boxplots para identificar outliers
for col in numeric_features:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f"Boxplot - {col}")
    plt.show()
```

#### Bivariate Analysis

```python
# Correlação entre features numéricas
plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_features].corr(), cmap='coolwarm', annot=False)
plt.title("Matriz de Correlação (features numéricas)")
plt.show()

# Relação entre features e target_regression
if 'target_regression' in df.columns:
    correlations = df.corr(numeric_only=True)['target_regression'].sort_values(ascending=False)
    print("Correlação das features com target_regression:")
    display(correlations)

# Relação entre features e target_class (se for categórica)
if 'target_class' in df.columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(x='target_class', data=df, palette='viridis')
    plt.title("Distribuição da variável target_class")
    plt.show()

    # Exemplo: média de uma feature por classe
    feature_example = numeric_features[0]
    plt.figure(figsize=(8, 5))
    sns.boxplot(x='target_class', y=feature_example, data=df)
    plt.title(f"{feature_example} vs target_class")
    plt.show()
```
