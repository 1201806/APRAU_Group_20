# Analise

```python
dups = df.duplicated().sum()
print(f"Duplicados encontrados: {dups}")
if dups > 0:
    df = df.drop_duplicates().reset_index(drop=True)
    print("Duplicados removidos.")
else:
    print("Sem duplicados exatos.")
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = [c for c in df.columns if c not in numeric_cols + ['y']]

# DiagnÃ³stico de valores missing
missing_percent = (df.isna().mean() * 100).sort_values(ascending=False)
if missing_percent.sum() > 0:
    display(missing_percent)
else:
    print("Sem valores NA diretos")
```

```python
def count_outliers_iqr(series, k=1.5):
    q1, q3 = series.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return ((series < lower) | (series > upper)).sum()

outlier_counts = {col: count_outliers_iqr(df[col]) for col in numeric_cols}
outlier_df = pd.DataFrame.from_dict(outlier_counts, orient='index', columns=['# Outliers'])
outlier_df['Percentual (%)'] = (outlier_df['# Outliers'] / len(df) * 100).round(2)
display(outlier_df.sort_values(by='Percentual (%)', ascending=False).head(10))
```

```python
def cramers_v_corrected(x, y):
    """CramÃ©r's V corrigido entre duas variÃ¡veis categÃ³ricas."""
    tbl = pd.crosstab(x, y)
    if tbl.size == 0 or tbl.shape[0] < 2 or tbl.shape[1] < 2:
        return np.nan
    chi2 = chi2_contingency(tbl, correction=False)[0]
    n = tbl.to_numpy().sum()
    phi2 = chi2 / n
    r, k = tbl.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2) / (n - 1)
    kcorr = k - ((k - 1)**2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

# DistribuiÃ§Ãµes categÃ³ricas
for col in categorical_cols:
    plt.figure(figsize=(7, 3))
    sns.countplot(x=df[col], order=df[col].value_counts().index)
    plt.title(f"DistribuiÃ§Ã£o de {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python

# --- DistribuiÃ§Ã£o do target_class ---
plt.figure(figsize=(5, 3))
sns.countplot(x=df["target_class"])
plt.title("DistribuiÃ§Ã£o da variÃ¡vel target_class")
plt.show()

# --- DistribuiÃ§Ã£o do target_regression ---
sns.histplot(df["target_regression"], kde=True)
plt.title("DistribuiÃ§Ã£o da variÃ¡vel target_regression")
plt.show()


# ===============================
# RelaÃ§Ã£o das categÃ³ricas com TARGET_CLASS (CramÃ©râ€™s V)
# ===============================
def cramers_v_corrected(x, y):
    """CramÃ©r's V corrigido entre duas variÃ¡veis categÃ³ricas."""
    tbl = pd.crosstab(x, y)
    if tbl.size == 0 or tbl.shape[0] < 2 or tbl.shape[1] < 2:
        return np.nan
    chi2 = chi2_contingency(tbl, correction=False)[0]
    n = tbl.to_numpy().sum()
    phi2 = chi2 / n
    r, k = tbl.shape
    phi2corr = max(0, phi2 - ((k - 1)*(r - 1)) / (n - 1))
    rcorr = r - ((r - 1)**2) / (n - 1)
    kcorr = k - ((k - 1)**2) / (n - 1)
    return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))


cat_corr_class = {}
for col in categorical_cols:
    cat_corr_class[col] = cramers_v_corrected(df[col], df["target_class"])

cat_corr_class = pd.Series(cat_corr_class).sort_values(ascending=False)
print("AssociaÃ§Ã£o CramÃ©r's V com target_class:")
display(cat_corr_class.head(10))

plt.figure(figsize=(6, 6))
sns.barplot(x=cat_corr_class.values[:10], y=cat_corr_class.index[:10], orient="h")
plt.title("Top 10 variÃ¡veis categÃ³ricas mais associadas a target_class")
plt.show()


# ===============================
# CorrelaÃ§Ã£o das numÃ©ricas com TARGET_CLASS (point-biserial)
# ===============================
num_corr_class = {}
y_class_codes = pd.Categorical(df["target_class"]).codes

for col in numeric_cols:
    try:
        corr, _ = pointbiserialr(y_class_codes, df[col])
        num_corr_class[col] = corr
    except Exception:
        num_corr_class[col] = np.nan

num_corr_class = pd.Series(num_corr_class).sort_values(ascending=False)
print("\nCorrelaÃ§Ã£o numÃ©rica com target_class:")
display(num_corr_class.head(10))

plt.figure(figsize=(6, 8))
sns.barplot(x=num_corr_class.values[:10], y=num_corr_class.index[:10], orient="h")
plt.title("Top 10 variÃ¡veis numÃ©ricas mais correlacionadas com target_class")
plt.show()


# ===============================
# CorrelaÃ§Ã£o das numÃ©ricas com TARGET_REGRESSION (Pearson)
# ===============================
if "target_regression" in df.columns:
    num_corr_reg = {}
    for col in numeric_cols:
        try:
            corr, _ = pearsonr(df[col], df["target_regression"])
            num_corr_reg[col] = corr
        except Exception:
            num_corr_reg[col] = np.nan

    num_corr_reg = pd.Series(num_corr_reg).sort_values(ascending=False)
    print("\nCorrelaÃ§Ã£o numÃ©rica com target_regression:")
    display(num_corr_reg.head(10))

    plt.figure(figsize=(6, 8))
    sns.barplot(x=num_corr_reg.values[:10], y=num_corr_reg.index[:10], orient="h")
    plt.title("focus_factor com target_regression")
    plt.show()
```
